```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay, classification_report

diabetes_df = pd.read_csv("diabetes.csv")

diabetes_df.head()
```

```python
from sklearn.model_selection import train_test_split

X = diabetes_df.drop('Outcome',axis=1)
y = diabetes_df['Outcome']

#split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
model =tree.DecisionTreeClassifier(criterion='entropy', max_depth=7, random_state=42,max_features=9,max_leaf_nodes=7)
```

```python
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
def arraynames(array):
    list = []
    string = ""
    x = 0
    for v in array:
        list.append(str(v.values()))
    while x <= int(len(list)-1):
        if x < int(len(list)-2):
            list[x] = str(list[x]) + ","
        elif x == int(len(list)-2):
            list[x] = str(list[x]) + " &"
        x += 1
    string = " ".join(list)
    return string
arraynames([ {'name': 'Niles'}, {'name': 'C.C.'}, {'name': 'Fran'} ])
```
