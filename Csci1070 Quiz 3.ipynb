```python
import numpy as np
import pandas as pd
import math
import statistics
from scipy import stats

belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
grouped_df = belize_df.groupby(["neighbourhood"])

price_mean = grouped_df[["price"]].mean().reset_index()
review_numb = grouped_df[["number_of_reviews"]].count().reset_index()
availibility = grouped_df[["availability_365"]].max().reset_index()

merged1 = pd.merge(review_numb, price_mean, on="neighbourhood")
merged2 = pd.merge(merged1, availibility, on="neighbourhood")
merged2
```
